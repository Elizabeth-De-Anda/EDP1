```python
import scipy
import numpy as np
import sympy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt

from sympy import simplify
from sympy import expand
```

### Problemas de Sturm-Liouville


### Ejercicio 1

$$  x'' + kx = 0 $$
$$ x(0) = 0 $$
$$ x(1) = 0 $$

```python
# Definimos las variables

x = sp.Function('x')
t = sp.symbols('t')
k = sp.symbols('k', positive = True)
C1, C2 = sp.symbols('C1, C2')
n = sp.symbols('n', positive = True)

# Resuelve la ecuación

ode = sp.Eq(x(t).diff(t, t) + k*x(t), 0)
sol = sp.dsolve(ode, x(t))

# Muestra la solución
sol
```

```python
# Aplicar las condiciones de frontera

conds = [sol.rhs.subs(t, 0), sol.rhs.subs(t, 1)]
sols = sp.solve(conds, (C1, C2))

sols
```

```python
# Como las soluciones de C1 = C2 = 0 es la trivial, tomamos el caso en que C2 = 0 y sen(√kt) con k = π^2 n^2

# Obtenemos los valores para k
val = sp.solve(sp.sin(sp.sqrt(k)), k)
val
```

```python
# Se obtiene la solución no trivial

sol_1 = sol.subs(C2, 0)
sol_2 = sol_1.subs(k, n**2 * sp.pi**2)

# Muestra la solcuión
sol_2
```

```python
#Gráficamos para algunas soluciones
# Para n = 8

n = 8
C1 = 1


xsol = C1 * sp.sin(sp.pi * n * t)
funct = sp.lambdify(t, xsol.subs(sp.pi, np.pi), 'numpy')

# Evaluar
tt = np.linspace(0, 1, 100)
xt = funct(tt)

# Graficar
plt.plot(tt, xt)
plt.xlabel('t')
plt.ylabel('x')
plt.title('Gráfica de x(t) = C1 sen(πnt)')
plt.grid()
plt.show()
```

### Ejercicio 2

$$ x'' + kx = 0 $$
$$ x'(0) = 0 $$
$$ x'(1) = 0 $$

```python
# Definimos las variables

x = sp.Function('x')
t = sp.symbols('t')
k = sp.symbols('k', positive = True)
C1, C2 = sp.symbols('C1, C2')
n = sp.symbols('n', positive = True)

# Resuelve la ecuación

ode = sp.Eq(x(t).diff(t, t) + k*x(t), 0)
sol = sp.dsolve(ode, x(t))

# Muestra la solución
sol
```

```python
 # Calcula la derivada de la solución

sold = sol.rhs.diff(t)

# Muestra la derivada
sold
```

```python
# Aplica las condiciones de frontera

conds = [sol.rhs.subs(t, 0), sold.subs(t, 0), sol.rhs.subs(t, 1), sold.subs(t, 1)]
sols = sp.solve(conds, (C1, C2))

sols
```

```python
# Como C1 = C2 = 0 es la solución trivial, ahora tomamos el caso en que C2 = 0 y k = (nπ)^2

# Se obtiene la solución no trivial

sol_1 = sol.subs(C1, 0)
sol_2 = sol_1.subs(k, (n*sp.pi)**2)

# Muestra la solcuión
sol_2
```

```python
#Gráficamos para algunas soluciones
# Para n = 8

n = 8
C2 = 1


xsol = C2 * sp.sin(sp.pi * n * t)
funct = sp.lambdify(t, xsol.subs(sp.pi, np.pi), 'numpy')

# Evaluar
tt = np.linspace(0, 1, 100)
xt = funct(tt)

# Graficar
plt.plot(tt, xt)
plt.xlabel('t')
plt.ylabel('x')
plt.title('Gráfica de x(t) = C2 cos(πnt)')
plt.grid()
plt.show()
```
