```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.1)
y = x*np.cos(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lab DLS')
plt.show()
```
